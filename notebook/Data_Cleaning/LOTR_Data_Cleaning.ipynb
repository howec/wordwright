```python
import pandas as pd
import numpy as np
import re
import pickle
```

```python
df = pd.read_csv('../Mental_Health_Hackathon2020/LordOfTheRings/LordOfTheRings_Dataframe.csv', encoding = "utf-8")
```

```python
df
```

```python
LOTR_text = df['dialog'].astype(str).values.tolist()
LOTR_text
```

```python
# clean text - model's focus on letters only not including punctuations 
# replace period with a space 
LOTR_text = [re.sub(r'[ ?.]', " ", i) for i in LOTR_text] 
# remove other punctuations
LOTR_text = [re.sub(r'[^\w\s]', "", i) for i in LOTR_text] 
# remove \xa0
LOTR_text = [re.sub(r'\\b \xa0\\b', " ", i) for i in LOTR_text]
# remove extra spaces
LOTR_text = [re.sub(r'\s+', " ", i) for i in LOTR_text]
```

```python
LOTR_text
```

```python
LOTR_text = ' '.join(LOTR_text)
```

```python
LOTR_text
```

```python
LOTR_list = []
LOTR_list += [LOTR_text] 
```

```python
LOTR_list
```

```python
with open('LOTR_list.pkl', 'wb') as f:
    pickle.dump(LOTR_list, f)
```
